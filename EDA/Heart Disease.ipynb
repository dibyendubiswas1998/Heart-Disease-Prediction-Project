```python
# Check the multicollinearity factor

corr = data.corr(method='spearman')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
data.columns
```

```python
# decting outliers

for column in data:
    df = data.copy()
    sns.boxplot(df[column])
    plt.title(column)
    plt.figure(figsize=(10,2))
```

```python
data.shape
```

```python
# Handle the outliers

# trestbps columns

IQR = data['trestbps'].quantile(0.75) - data['trestbps'].quantile(0.25)
lower = data['trestbps'].quantile(0.25) - (IQR * 1.5)
upper = data['trestbps'].quantile(0.75) + (IQR * 1.5)
print(f"IQR: {IQR},  lower: {lower},  upper: {upper}")


data.loc[df['trestbps']>=upper, 'trestbps']=upper
data.loc[df['trestbps']<=lower, 'trestbps']=lower
```
